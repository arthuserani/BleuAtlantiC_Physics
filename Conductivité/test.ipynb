```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit   
import itertools
import pandas as pd
```

```python
2,363	2,679
2,35	2,698
2,346	2,697
2,351	2,685
2,354	2,704
2,36	2,645
```

```python
m_avant_flush = [0, 2.679-2.363, 2.698-2.35, 2.697-2.346]
m_après_flush= [0, 2.685-2.351, 2.704-2.354, 2.645-2.36]
m_avant_flush = [m_avant_flush[0], m_avant_flush[0]+m_avant_flush[1], m_avant_flush[0]+m_avant_flush[1]+m_avant_flush[2], m_avant_flush[0]+m_avant_flush[1]+m_avant_flush[2]+m_avant_flush[3]]
m_après_flush = [m_après_flush[0], m_après_flush[0]+m_après_flush[1], m_après_flush[0]+m_après_flush[1]+m_après_flush[2], m_après_flush[0]+m_après_flush[1]+m_après_flush[2]+m_après_flush[3]]
```

```python
minutes = np.array([0, 1, 2, 3]) * 5

# Fonction affine pour l'ajustement
def func(x, a, b):
    return a * x + b

# Création de la figure avec 8 sous-graphiques (2 lignes, 4 colonnes)
fig = plt.figure()
ax = fig.subplots(1,1)
ydata = m_avant_flush
popt, _ = curve_fit(func, minutes, ydata)
    
    # Calcul de R²
residuals = ydata - func(minutes, *popt)
ss_res = np.sum(residuals**2)
ss_tot = np.sum((ydata - np.mean(ydata))**2)
r_squared = 1 - (ss_res / ss_tot)

ax.plot(minutes, ydata, 'o', color='blue', label='Data')
ax.plot(minutes, func(minutes, *popt), 'r-', label=f'Fit, $\\mathbf{{R^2={r_squared:.3f}}}$')
ax.set_ylabel('Mass (g)')
ax.set_xlabel('Time (min)')
ax.legend()
    
    # Affichage du débit
ax.text(0.05, 0.95, f'Q = {popt[0]:.3f} g/min', transform=ax.transAxes,
fontsize=10, color='k', fontweight='bold')
ax.text(0.05, 0.85, f'Q = {popt[0]*1000/60:.3f} mg/s', transform=ax.transAxes,
fontsize=10, color='k', fontweight='bold')

plt.tight_layout()
plt.show()
```
